```python
from nltk.tokenize import word_tokenize ,sent_tokenize
```

```python
input = "Hello Mr. Smith, how are you doing today? The weather is great, and Python is awesome. The sky is pinkish-blue. You shouldn't eat cardboard."
output1 = word_tokenize(input)
print(output1)
output2 = sent_tokenize(input)          
print(output2)
```

```python
import spacy 
nlp = spacy.load("en_core_web_sm")
input = "This is a sentence written by me "
doc = nlp(input)
```
